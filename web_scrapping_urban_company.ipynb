# Scrapping the Names, Dates, Ratings, Comments/Services.
- Name contains name of customers given ratings.
- Dates contains on which dates.
- ratings cotain how many ratings
- comments contain no comments but the service name and city as well as country name in single text.

```python
names = []
dates = []
rates = []
comments = []

for i in range(1, 1000): #Loop for multiple pages.
    
    try:
        urls = f"https://www.urbancompany.com/reviews?p={i}" 
        ct = requests.get(urls)
        
    except Exception as e:
        print(e)
    
    
    soup = BeautifulSoup(ct.content)
    coms = soup.find_all("div", class_ ="ReviewPageReview__topInfo--3ibra")


    for com in coms:
        name = com.find("span", class_ = "ellipsis").text
        date = com.find("span", class_ = "ReviewPageReview__reviewDate--3tC00").text    
        ratings = com.find("div", class_ = "StarRatingSEO__ratingBox--3MqzX StarRatingSEO__ratingBoxSmall--4Nhn8").text
        reviews = com.find("div", class_ = "ReviewPageReview__seoLinkDiv--2RnCy")
        
        if reviews is None:
            comments.append("None")
        else:
            comments.append(reviews.text)
            

        names.append(name)
        dates.append(date)
        rates.append(ratings)
        
        print(name)
        print(date)
        print(ratings)
        print(reviews)
        print("="*50)
```

```python
print(len(names))
print(len(dates))
print(len(rates))
print(len(comments))
```

# Spliting the comments/reviews text into places and services.
- As there are no comments but there are services with the city and country written.

```python
Services = []
Places = []

for i in comments:
    
    if i != "None":
        services,place=i.split(" in ")
        Services.append(services)
        Places.append(place)
        
    elif i == "None":
        Services.append("None")
        Places.append("None")
```

```python
print(len(Services))
print(len(Places))
```

# Spiltting into country and city from places list.
- after sppliting into services and places we got the places list which contains city and country so splitted the as well.

```python
Country = []
City = []
for i in Places:
    
    if i == "None":
        City.append("None")
        Country.append("None")
        
    elif i != "None":
        city,country=i.split(", ")
        Country.append(country)
        City.append(city)
```

```python
print(len(Country))
print(len(City))
```

# Creating DataFrame
- DataFarame contains Dates,name,ratings,city,country and comments in it.

```python
df = pd.DataFrame()
```

```python
df["Dates"] = dates
df["Names"] = names
df["Ratings"] = rates
df["Services"] = Services
df["City"] = City
df["Counrty"] = Country
df["Comments"] = comments
```
